```
# Import required libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

from vortx import EarthMemorySystem
from vortx.utils import setup_environment
from vortx.viz import MemoryVisualizer
```

## 1. System Setup

First, let's set up the Earth Memory System with basic configurations.

```
# Configure environment
setup_environment(
    api_key=os.getenv("VORTX_API_KEY", "demo-key"),
    log_level="INFO"
)

# Initialize system
system = EarthMemorySystem(
    memory_config={
        "architecture": "hierarchical",
        "compression": "adaptive"
    }
)

print("System initialized successfully!")
```

## 2. Loading Sample Data

Let's generate some sample Earth observation data for demonstration.

```
def generate_sample_data(n_samples=1000):
    """Generate synthetic Earth observation data."""
    start_date = datetime(2024, 1, 1)
    
    data = {
        "timestamp": [start_date + timedelta(hours=i) for i in range(n_samples)],
        "latitude": np.random.uniform(30, 45, n_samples),
        "longitude": np.random.uniform(-120, -100, n_samples),
        "temperature": np.random.normal(25, 5, n_samples),
        "precipitation": np.random.exponential(2, n_samples),
        "cloud_cover": np.random.uniform(0, 1, n_samples)
    }
    
    return pd.DataFrame(data)

# Generate and display sample data
sample_data = generate_sample_data()
print("Sample data shape:", sample_data.shape)
sample_data.head()
```

## 3. Creating Earth Memories

Now, let's create Earth memories from our sample data.

```
# Create memory from sample data
memory = system.create_memory(
    data=sample_data,
    memory_params={
        "spatial_resolution": "1km",
        "temporal_resolution": "1h",
        "compression_level": "medium"
    }
)

print("Memory created successfully!")
print(f"Memory size: {memory.size_mb:.2f} MB")
```

## 4. Basic Operations

Let's perform some basic operations on our Earth memory.

```
# Query memory
query_results = memory.query(
    bbox=[-115, 35, -105, 40],  # [min_lon, min_lat, max_lon, max_lat]
    time_range=[datetime(2024, 1, 1), datetime(2024, 1, 7)],
    variables=["temperature", "precipitation"]
)

print("Query results shape:", query_results.shape)
query_results.head()
```

## 5. Visualization

Finally, let's visualize our Earth memory data.

```
# Initialize visualizer
visualizer = MemoryVisualizer()

# Create visualization
fig = visualizer.plot_memory(
    memory=memory,
    variable="temperature",
    time=datetime(2024, 1, 1),
    plot_type="heatmap"
)

plt.show()
```

## Cleanup

Always clean up system resources when done.

```
# Cleanup
system.cleanup()
print("Cleanup complete!")
```
