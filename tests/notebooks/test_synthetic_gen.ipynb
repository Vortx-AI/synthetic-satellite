```python
import logging
from pathlib import Path
from src.synthetic.generator import (
    SyntheticGenerator,
    TerrainParams,
    AtmosphericParams,
    SpectralParams
)

def main():
    # Configure logging
    logging.basicConfig(level=logging.INFO)
    
    # Define configuration for the generator
    config = {
        "image_size": (512, 512),       # Reduced size for quicker testing
        "num_bands": 4,                  # RGBN (Red, Green, Blue, Near-Infrared)
        "resolution": 10.0,              # meters per pixel
        "random_seed": 123               # Seed for reproducibility
    }
    
    # Initialize the SyntheticGenerator with the configuration
    generator = SyntheticGenerator(config=config)
    
    # Define terrain parameters
    terrain_params = TerrainParams(
        elevation_scale=800.0,
        roughness=0.4,
        persistence=0.6,
        octaves=4,
        base_frequency=0.03
    )
    
    # Define atmospheric parameters
    atmospheric_params = AtmosphericParams(
        cloud_coverage=0.25,
        haze_intensity=0.15,
        aerosol_density=0.12,
        shadow_intensity=0.35
    )
    
    # Define spectral parameters
    spectral_params = SpectralParams(
        band_correlations={
            "red": 1.0,
            "green": 0.9,
            "blue": 0.7,
            "nir": 0.5
        },
        spectral_noise=0.05,
        sensor_gain=1.0,
        quantization_bits=12
    )
    
    # Specify the output path for the synthetic GeoTIFF image
    output_path = Path("output/synthetic_scene_test.tif")
    
    # Ensure the output directory exists
    output_path.parent.mkdir(parents=True, exist_ok=True)
    
    # Generate the synthetic scene
    try:
        scene_metadata = generator.generate_scene(
            output_path=output_path,
            terrain_params=terrain_params,
            atmospheric_params=atmospheric_params,
            spectral_params=spectral_params
        )
        logging.info(f"Synthetic scene generated successfully at {output_path}")
        logging.info(f"Scene Metadata: {scene_metadata}")
    except Exception as e:
        logging.error(f"Failed to generate synthetic scene: {e}")


main()
```
